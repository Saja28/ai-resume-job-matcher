```python
match_report = {
    "company": "hays",
    "role": "machine learning engineer",
    "seniority": "mid",
    "match_percent": "60.95",
    "matched_skills": [
        "python",
        "google_cloud_platform",
        "machine_learning",
        "feature_engineering",
        "model_evaluation",
        "model_training"
    ],
    "important_missing_skills": [
        "pytorch",
        "recommendation_systems",
        "model_deployment",
        "airflow",
        "kubernetes",
        "apache_spark",
        "deep_learning",
        "google_bigquery",
        "inference_pipelines",
        "training_pipelines",
        "search_ranking_models",
        "evaluation_pipelines"
    ]
}
```

```python
prompt = f"""
You are a senior machine learning career coach.

Analyze this job match report and provide:

1. Candidate strengths
2. Missing skills
3. Recommended projects
4. Resume improvements
5. Learning roadmap

Match Report:

{match_report}
"""
```

```python
def generate_fallback_recommendation(match_report):
    print(f"Role: {match_report['role']}")
    print(f"Company: {match_report['company']}")
    print(f"Match Score: {match_report['match_percent']}%")

    print("\nStrengths:")
    for skill in match_report["matched_skills"]:
        print(f"✅ You already show {skill}.")

    print("\nMissing Skills:")
    for skill in match_report["important_missing_skills"]:
        print(f"❌ Add evidence for {skill}.")

    print("\nRecommended Projects:")
    print("1. Build a PyTorch-based skill classification model.")
    print("2. Build a recommendation system using job and resume embeddings.")
    print("3. Build a small MLOps pipeline with model deployment and monitoring.")

generate_fallback_recommendation(match_report)
```

```python
def generate_fallback_recommendation(match_report):
    strengths = match_report["matched_skills"]
    missing = match_report["important_missing_skills"]

    recommendation = f"""
AI Resume Recommendation Report

Role: {match_report['role']}
Company: {match_report['company']}
Match Score: {match_report['match_percent']}%

Strengths:
{chr(10).join([f"- {skill}" for skill in strengths])}

Priority Missing Skills:
{chr(10).join([f"- {skill}" for skill in missing[:6]])}

Recommended Projects:
1. PyTorch Skill Classification Model
2. Resume-Job Recommendation System
3. MLOps Pipeline with Deployment and Monitoring

Next Learning Focus:
- PyTorch
- Model Deployment
- Airflow
- Kubernetes
- Recommendation Systems
"""
    return recommendation

fallback_report = generate_fallback_recommendation(match_report)

print(fallback_report)
```

```python
with open("gemini_fallback_report.txt", "w") as file:
    file.write(fallback_report)
```
